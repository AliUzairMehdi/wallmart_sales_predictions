```python
import csv
```

```python
def read_sales_data():
    with open('./raw_sales_data.csv','r') as csv_file:
        csv_reader= csv.reader(csv_file)
        header=next(csv_reader)
        print(f"Header: {header}")

        for row in csv_reader:
            print(row)
```

```python
df=df.drop_duplicates(subset=['OrderID','CustomerName'],keep='first')
```

```python
df[:50]
```

```python
df['CustomerName']=df['CustomerName'].fillna('Unknown')
```

```python
df[60:120]
```

```python
df['Category']=df['Category'].str.lower().str.strip()
df['Category']=df['Category'].replace({'fash':'Fashion','elec':'Electronics'})
```

```python
df[60:120]
```

```python
working_df=df
working_df
```

```python
working_df['Quantity']=working_df['Quantity'].dropna()
```

```python
working_df
```

```python
working_df=working_df[working_df['Quantity']>=0]
```

```python
working_df
```

```python
working_df=working_df[working_df['UnitPrice']>=0]
```

```python
working_df
```

```python
working_df['PaymentMethod']=working_df['PaymentMethod'].fillna("Unknown")
```

```python
import sqlite3
```

```python
try:
    conn = sqlite3.connect('sales_data.db')
    print("Connection to SQLite DB successful")
except sqlite3.Error as e:
    print(f"Error connecting to SQLite DB: {e}")
```

```python
cursor=conn.cursor()
cursor.execute("""
CREATE TABLE IF NOT EXISTS orders (
    OrderID INTEGER,
    CustomerName TEXT,
    Product TEXT,
    Category TEXT,
    Quantity FLOAT,
    UnitPrice FLOAT,
    OrderDate DATE,
    PaymentMethod TEXT
)
""")
```

```python

if conn:
    conn.close()
    print("SQLite connection closed")
```
